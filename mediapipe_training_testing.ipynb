```python
from mediapipe_analysis import *
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, LeakyReLU, Conv2D
from keras.optimizers import RMSprop
from matplotlib.pyplot import imshow
import numpy as np
```

```python
# load data
x_train, x_test, y_train, y_test = loaddata_mediapipe(
    os.path.join(os.path.realpath("./"), "1"),
    os.path.join(os.path.realpath("./"), "2"),
    os.path.join(os.path.realpath("./"), "3"),
    rand_state=777
)
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
y_train = keras.utils.to_categorical(y_train, 5)
y_test = keras.utils.to_categorical(y_test, 5)
print(x_train.shape, y_train.shape, x_test.shape, y_test.shape)
```

```python
# create model
if os.path.isfile("model_nn_mediapipe.h5"):
    model = keras.models.load_model("model_nn_mediapipe.h5")
else:
    model = Sequential()
    model.add(Dense(20, activation=LeakyReLU(), input_shape=(20,)))
    model.add(Dense(20, activation=LeakyReLU()))
    model.add(Dense(5, activation='softmax'))
    model.compile(loss='categorical_crossentropy',
                  optimizer=RMSprop(),
                  metrics=['accuracy'])
```

```python
# training
batch_size = 256
epochs = 256
history = model.fit(x_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_data=(x_test, y_test))
```

```python
# saving
model.save("model_nn_mediapipe.h5")
```

```python
# testing
img = cv2.imread("./2/data/0147.png")
inputs = get_nn_input_from_img(img)
print(np.argmax(model.predict([inputs]), axis=1))
```

```python
# show metric
from sklearn import metrics
y_pred = model.predict(x_test)
for i_entry in range(0, len(y_pred)):
    max_value = max(y_pred[i_entry])
    for i in range(0, len(y_pred[i_entry])):
        if y_pred[i_entry][i] == max_value:
            y_pred[i_entry][i] = 1
        else:
            y_pred[i_entry][i] = 0
print(y_test.shape, y_pred.shape)
print(metrics.classification_report(keras.utils.to_categorical(y_test, 5), y_pred))
```
